```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from astropy.io.fits import getdata
import matplotlib.colors as colors
```

```python
data = getdata('output.fits', header=False)
```

```python
mind = data['dec'].min()
maxd = data['dec'].max()
minr = data['ra'].min()
maxr = data['ra'].max()
bind = 2*(maxd - mind)
binr = 2*(maxr - minr)
```

```python
gridx = np.linspace(minr,maxr,binr)
gridy = np.linspace(mind,maxd,bind)#min/max from data
```

```python
b = ((data['r'] > 20.0) & (data['r'] <= 20.66))
g = ((data['r'] > 20.66) & (data['r'] <= 21.33))
r = ((data['r'] > 21.33) & (data['r'] <= 22.0))
ran=[[mind,maxd],[minr,maxr]]
#ran=[[0,360],[-90,90]]
bins=[bind,binr]
#bins=[gridx,gridy]
```

```python
red, x, y = np.histogram2d(data['dec'][r],data['ra'][r],bins=bins,range=ran)
green, x, y = np.histogram2d(data['dec'][g],data['ra'][g],bins=bins,range=ran)
blue, x, y = np.histogram2d(data['dec'][b],data['ra'][b],bins=bins,range=ran)
```

```python
rmask = np.where((red > 82.5))
gmask = np.where((green > 76))
bmask = np.where((blue > 66.75))
red[rmask] = 0
green[gmask] = 0#np.nan
blue[bmask] = 0#np.nan
```

```python
rmask = np.where((red < 25.5))
gmask = np.where((green < 26.75))
bmask = np.where((blue < 24.5))
red[rmask] = 0
green[gmask] = 0#np.nan
blue[bmask] = 0#np.nan
```

```python
print red.max()
print red.mean()
print green.max()
print green.mean()
print blue.max()
print blue.mean()
#print red.tolist()
```

```python
extent = [y[-1],y[0],x[0],x[-1]]
print extent
#longitude_top_left,longitude_top_right,latitude_bottom_left,latitude_top_left
print np.size(red[np.nonzero(red)]) + np.size(blue[np.nonzero(blue)]) + np.size(green[np.nonzero(green)])
```

```python
rgb = np.zeros((bind,binr,3),dtype='uint8')
rgb[:,:,0] = red / (red.max() / 255)
rgb[:,:,1] = green / (green.max() / 255)
rgb[:,:,2] = blue / (blue.max() / 255)
stream = matplotlib.image.FigureImage(rgb)
```

```python
print rgb[:,:,0].max()
print rgb[:,:,0].mean()
print rgb[:,:,1].max()
print rgb[:,:,1].mean()
print rgb[:,:,2].max()
print rgb[:,:,2].mean()
```

```python
rgb = np.flipud(rgb)
rgb = np.fliplr(rgb)
```

```python
fig,ax=plt.subplots(figsize=(10,10))
im = ax.imshow(rgb,cmap=plt.cm.gray,extent=extent,interpolation='nearest',origin='upper')

```
